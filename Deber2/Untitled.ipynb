```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from scipy import misc
from PIL import Image

def cargar_imagen():
    imagen_cachorro_abrir = Image.open('cachorro.jpg')
    imagen_cachorro = np.array(imagen_cachorro_abrir)
    return imagen_cachorro

imagen_cachorro=cargar_imagen()

class Roompecabezas:
    img1=None
    img2=None
    img3=None
    img4=None
    img5=None
    img6=None
    imagen_ordenada_resultado=np.zeros([600, 900, 3])
    #imagen_final=None
    #imagen_final2=None
    
    def __init__(self, imagen):
        arr7, arr8, arr9  = np.hsplit(imagen_cachorro,3)
        arr1, arr4  = np.vsplit(arr7,2)
        arr2, arr5  = np.vsplit(arr8,2)
        arr3, arr6  = np.vsplit(arr9,2)
        
        self.img1 = arr1
        self.img2 = arr2
        self.img3 = arr3
        self.img4 = arr4
        self.img5 = arr5
        self.img6 = arr6 
        #self.imagen_ordenada_resultado= np.zeros([600, 900, 3])
        
    def desordenar_imagen(self):
        res_img1 = Image.fromarray(self.img1)
        res_img2 = Image.fromarray(self.img2)
        res_img3 = Image.fromarray(self.img3)
        res_img4 = Image.fromarray(self.img4)
        res_img5 = Image.fromarray(self.img5)
        res_img6 = Image.fromarray(self.img6)
        final = Image.new("RGB",(1100,700),"white")
    
        final.paste(res_img6, (0,0))
        final.paste(res_img4, (400,0))
        final.paste(res_img1, (800,0))
        
        final.paste(res_img3, (0,400))
        final.paste(res_img2, (400,400))
        final.paste(res_img5, (800,400))
        imagen_cachorro_desordenada = np.array(final)
        
        
        return imagen_cachorro_desordenada      
    def mostrar_imagen(self,imagen_cachorro):
        #imagen_cachorro_abrir = Image.open('cachorro.jpg')
        #imagen_cachorro = np.array(imagen_cachorro_abrir)
        plt.imshow(imagen_cachorro)
        plt.show
    def inicio(self):
        img_desordenada=desordenar_imagen(imagen_cachorro)
        inicio="Roompecabezas".capitalize()     
        print(inicio.center(50, "=") )    
        mostrar_imagen(imagen_cachorro)
        plt.title("Imagen ordenada")
        plt.figure()
        plt.title("Imagen desordenada")  
        mostrar_imagen(img_desordenada)
    
    def ordenar_imagen(self):
        #print(imagen_ordenada_resultado)
        numero_imagen = int(input("Ingrese el nmero de la imagen: "))
        posicion_imagen= int(input("Ingrese la posicion de la imagen: "))
        
        #print(self.imagen_ordenada_resultado)
        
        if (self.imagen_ordenada_resultado.any() == 0):
            #print("todos ceros")
            final = Image.new("RGB",(900,600),"white")
        else:
            #print("numeros")
            final= Image.fromarray(self.imagen_ordenada_resultado) 
            
        
        if(numero_imagen==1):
            res_img = Image.fromarray(self.img6)
        if(numero_imagen==2):
            res_img = Image.fromarray(self.img4)
        if(numero_imagen==3):
            res_img = Image.fromarray(self.img1)
        if(numero_imagen==4):
            res_img = Image.fromarray(self.img3)
        if(numero_imagen==5):
            res_img = Image.fromarray(self.img2)
        if(numero_imagen==6):
            res_img = Image.fromarray(self.img5)
    
        if(posicion_imagen==1):
            final.paste(res_img, (0,0))
        if(posicion_imagen==2):
            final.paste(res_img, (300,0))
        if(posicion_imagen==3):
            final.paste(res_img, (600,0))        
        if(posicion_imagen==4):
            final.paste(res_img, (0,300))
        if(posicion_imagen==5):
            final.paste(res_img, (300,300))
        if(posicion_imagen==6):
            final.paste(res_img, (600,300))
        
        self.imagen_ordenada_resultado= np.array(final)
        return self.imagen_ordenada_resultado
        

        
rpc = Roompecabezas(imagen_cachorro)
  
img_desordenada=rpc.desordenar_imagen()
inicio="Roompecabezas".capitalize()     
print(inicio.center(50, "=") )    
rpc.mostrar_imagen(imagen_cachorro)
plt.title("Imagen ordenada")
plt.figure()
plt.title("Imagen desordenada")  
rpc.mostrar_imagen(img_desordenada)    
```
