```python
class Profesor:
    nombre=None
    cedula=None
    direccion=None
    telefono=None   
    def __str__(self):
        return f"Nombre:{self.nombre} Cédula:{self.cedula} Dirección:{self.direccion} Teléfono:{self.telefono}"
class Curso:
    codigo=None
    nombre=None   
    fecha_inicio=None
    fecha_fin=None
    profesor=None     
def listar_profesor():
    try:
        
        path="./profesor.txt"
        archivo_abierto=open(path)
        contenido = archivo_abierto.readlines()
        print("==Lista de profesores==")
        #print("Nombre\t\t\t\tCedula\t\t\tDireccion\t\t\tTelefono") 
        for linea in contenido:          
            if(linea !="\n"):
                dividirLista=linea.split(",")
                print(f"{dividirLista[0]}\t{dividirLista[1]}\t{dividirLista[2]}\t{dividirLista[3]}")      
    
        archivo_abierto.close()
    except Exception as error:
        print(error)    
def ingresar_profesor():
    try:       
        print("==Ingresar Profesor==")
        profesor = Profesor()
        profesor.nombre= input ("Ingrese el nombre: ")
        profesor.cedula= input ("Ingrese la cédula: ")
        profesor.direccion= input ("Ingrese la dirección: ")
        profesor.telefono= input ("Ingrese el teléfono: ")
        path="./profesor.txt"
        archivo_escritura_abierto=open(path, mode="a")       
        archivo_escritura_abierto.write(f"{profesor.nombre},{profesor.cedula},{profesor.direccion},{profesor.telefono}\n")
        archivo_escritura_abierto.close()
        print("Exito!!!")
    except Exception as error:
        print("error" + error)       
def modificar_profesor():
    try:
        path="./profesor.txt"
        listar_profesor()    
        cedula_profesor_modificar= input ("Ingrese la cédula del profesor a modificar: ")          
        archivo_abierto = open(path, "r")
        contenido = archivo_abierto.readlines()          
        archivo_abierto = open(path, "w")                         
        for linea in contenido:                     
            dividir_lista = linea.split(",")
            if (cedula_profesor_modificar != dividir_lista[1]): #si no encontro la cedula en el registro
                archivo_abierto.write(linea)
            else:                
                nombre= input ("Ingrese el nombre: ")
                cedula= dividir_lista[1]
                direccion= input ("Ingrese la dirección: ")
                telefono= input ("Ingrese el teléfono: ")
                if(nombre==""):
                    nombre=dividir_lista[0]
                if(direccion==""):
                    direccion=dividir_lista[2]
                if(telefono==""):
                    telefono=dividir_lista[3]
                
                archivo_abierto.write(f"{nombre},{cedula},{direccion},{telefono}\n")         
        print("Exito!!")
        archivo_abierto.close()              
    except Exception as error:
        print(error)    
def eliminar_profesor():
    try:  
        listar_profesor()       
        path="./profesor.txt"
        cedula_profesor_eliminar= input ("Ingrese la cédula del profesor a eliminar: ")         
        archivo_abierto = open(path, "r")
        contenido = archivo_abierto.readlines()
        archivo_abierto = open(path, "w")                        
        for linea in contenido:                     
            dividir_lista = linea.split(",")
            if (cedula_profesor_eliminar != dividir_lista[1]): #si no encontro la cedula en el registro
                archivo_abierto.write(linea)                               
        archivo_abierto.close()
        print("Exito!")
    except Exception as error:
        print(error)
def listar_curso():
    try:
        
        path="./curso.txt"
        archivo_abierto=open(path)
        contenido = archivo_abierto.readlines()
        print("Lista de cursos")
        #print("Nombre\t\t\t\tCedula\t\t\tDireccion\t\t\tTelefono") 
        for linea in contenido:          
            if(linea != "\n"):
                dividir_lista=linea.split(",")
                print(f"{dividir_lista[0]}\t{dividir_lista[1]}\t{dividir_lista[2]}\t{dividir_lista[3]}\t{dividir_lista[4]}")      
    
        archivo_abierto.close()
    except Exception as error:
        print(error)    
def ingresar_curso():
    try:        
        curso = Curso()
        curso.codigo= input ("Ingrese el código: ")
        curso.nombre= input ("Ingrese el nombre: ")
        curso.fecha_inicio= input ("Ingrese la fecha inicio: ")
        curso.fecha_fin= input ("Ingrese la fecha fin: ")
        path="./profesor.txt"
        archivo_abierto=open(path)
        contenido = archivo_abierto.readlines()        
        listar_profesor()
        curso.profesor=input("Ingrese la cédula de un profesor: ")        
        path="./curso.txt"
        archivo_escritura_abierto=open(path, mode="a")       
        archivo_escritura_abierto.writelines(f"{curso.codigo},{curso.nombre},{curso.fecha_inicio},{curso.fecha_fin},{curso.profesor}\n")
        archivo_escritura_abierto.close()
        print("Exito!!!")
    except Exception as error:
        print("error" + error)       
def modificar_curso():
    try:
        
        listar_curso()
        path="./curso.txt"
        codigo_curso_modificar= input ("Ingrese el código del curso a modificar: ")        
        archivo_abierto = open(path, "r")
        contenido = archivo_abierto.readlines()          
        archivo_abierto = open(path, "w")                         
        for linea in contenido:
            dividir_lista = linea.split(",")            
            if (codigo_curso_modificar != dividir_lista[0]): #si no encontro el codigo en el registro
                archivo_abierto.write(linea)
            else:             
                nombre= input ("Ingrese el nombre: ")
                codigo= dividir_lista[0]
                fecha_inicio= input ("Ingrese la fecha inicio: ")
                fecha_fin= input ("Ingrese la fecha fin: ")
                listar_profesor()
                profesor= input("Ingrese el profesor: ")
                if(nombre==""):
                    nombre=dividir_lista[1]
                if(fecha_inicio==""):
                    fecha_inicio=dividir_lista[2]
                if(fecha_fin==""):
                    fecha_fin=dividir_lista[3]
                if(profesor==""):
                    profesor=dividir_lista[4]
                archivo_abierto.write(f"{codigo},{nombre},{fecha_inicio},{fecha_fin},{profesor}\n")
        print("Exito!!")
        archivo_abierto.close()              
    except Exception as error:
        print(error)   
def eliminar_curso():
    try:
        listar_curso() 
        path="./curso.txt"
        codigo_curso_eliminar= input ("Ingrese el código del curso a eliminar: ")         
        archivo_abierto = open(path, "r")
        contenido = archivo_abierto.readlines()
        archivo_abierto = open(path, "w")                        
        for linea in contenido:                     
            dividir_lista = linea.split(",")
            if (dividir_lista[0] != codigo_curso_eliminar): #si no encontro el codigo en el registro
                archivo_abierto.write(linea)                           
        archivo_abierto.close()
        print("Exito!")
    except Exception as error:
        print(error)
def ingresar_opcion(): 
    correcto=False
    num=0
    while(not correcto):
        try:
            num = int(input("Elige una opción: "))
            correcto=True
        except ValueError:
            print('Error, introduce un numero entero')     
    return num 
salir = False
opcion = 0
def menu_profesor():
    print ("1. Listar")
    print ("2. Ingresar")
    print ("3. Actualizar")
    print ("4. Eliminar")
    print ("5. Salir")        
    opcion_profesor = ingresar_opcion()
    if opcion_profesor == 1:
        listar_profesor()
    if opcion_profesor == 2:
        ingresar_profesor()
    if opcion_profesor == 3:
        modificar_profesor()
    if opcion_profesor == 4:
        eliminar_profesor()
    if opcion_profesor == 5:
        salir = True       
def menu_curso():
    print ("1. Listar")
    print ("2. Ingresar")
    print ("3. Actualizar")
    print ("4. Eliminar")
    print ("5. Salir")        
    opcion_curso = ingresar_opcion()
    if opcion_curso == 1:
        listar_curso()
    if opcion_curso == 2:
        ingresar_curso()
    if opcion_curso == 3:
        modificar_curso()
    if opcion_curso == 4:
        eliminar_curso()
    if opcion_curso == 5:
        salir = True
while not salir:
    inicio="Bienvenido a mi aplicacion".capitalize() 
    print(inicio.center(50, "=") )
    print ("1. Profesor")
    print ("2. Curso")
    print ("3. Salir")
    opcion = ingresar_opcion() 
    if opcion == 1:
        inicio_profesor="Profesor".capitalize() 
        print(inicio_profesor.center(50, "=") )
        menu_profesor()        
    elif opcion == 2:
        inicio_curso="Curso".capitalize() 
        print(inicio_curso.center(50, "=") )
        menu_curso()
    elif opcion == 3:
        salir = True
    else:
        print ("Introduce un numero entre 1 y 3")
print ("Fin")   
```
